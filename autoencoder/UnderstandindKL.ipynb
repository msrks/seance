# KL-divergence

## Discrete


$D_{KL}(P||Q) = \sum_i P_i \log(\frac{P_i}{Q_i}) $

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

points = 100
shift = 3


x = np.linspace(-10., 10., points)
p = np.exp(-x**2/2)
p /= np.sum(p)

q = np.exp(-(x-shift)**2/2)
q /= np.sum(q)

plt.plot(x, p)
plt.plot(x, q)

d = p * np.log(p/q)
plt.plot(x, d)

plt.legend(['p', 'q', 'd'])
print(sum(d))
```

## Continuous 

```python
points = 100
shift = 3

x = np.linspace(-10., 10., points)
p = np.exp(-x**2/2) 
p /= np.sum(p) * (20. / points) 

q = np.exp(-(x-shift)**2/2)
q /= np.sum(q) * (20. / points)

plt.plot(x, p)
plt.plot(x, q)

d = p * np.log(p/q)
plt.plot(x, d)

plt.legend(['p', 'q', 'd'])
print(sum(d * (20. / points)))
```

## Analytical

$$p(x) = \frac{1}{\sqrt{2\pi}}e^{-x^2/2}$$
$$q(x) = \frac{1}{\sqrt{2\pi}}e^{-(x-\mu)^2/2}$$

$$D_{KL}(p||q) = \int p(x) \log(\frac{p(x)}{q(x)}) dx = \int \frac{1}{\sqrt{2\pi}}e^{-x^2/2} * ((x-\mu)^2-x^2)/2 dx = $$
$$ = \int \frac{1}{\sqrt{2\pi}}e^{-x^2/2} * (-2\mu x + \mu^2)/2 dx = \mu^2/2$$
